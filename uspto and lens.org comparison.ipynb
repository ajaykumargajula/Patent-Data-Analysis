```python
import pandas as pd
import numpy as np
data=pd.read_csv("E:\\BABI\\Capstone\\Capstone\\Capstone_Legal Analytics\\Data\\AllPatents.csv",index_col=0)
data_samsung=data[data["Company"]=="Samsung"]
data_samsung.head()

#data_subset=data_samsung[["URL","Company","Lens ID","Application Number"]]
```

```python
data_subset=data_samsung[["URL","Company","Lens ID","Application Number"]]
data_application=data_subset["Application Number"].str.split(" ",n=3,expand=True)
```

```python
data_application.columns=["Country","Application_Number","Application_status"]
data_application.head()
```

```python
samsung=pd.concat([data_subset.reset_index(drop=True),data_application],axis=1)
samsung.head()
```
